```python
#Frequência de Cargos
frequencia_cargos = df['job_title'].value_counts()

plt.figure(figsize=(12,5))

frequencia_cargos.plot(kind="bar", color="g")
plt.title("Frequência Cargos")
plt.xlabel("Cargos")
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()
```

```python
#Frequència de Experiència
frequencia_experiencia = df['experience_level'].value_counts()

plt.figure(figsize=(12,5))

frequencia_experiencia.plot(kind="bar", color="g")
plt.title("Frequência Experiência")
plt.xlabel("Experiência")
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()
```

```python
#Frequência de Tamanho de Empresa
frequencia_tam_empresa = df['company_size'].value_counts()

plt.figure(figsize=(12,5))

frequencia_tam_empresa.plot(kind="bar", color="g")
plt.title("Frequência de Tamanho de Empresa")
plt.xlabel("Tamanho empresa")
plt.ylabel("Frequência")

plt.tight_layout()
plt.show()
```

```python
# Medidas de tendência central
mean_salary = df['salary_in_usd'].mean()
median_salary = df['salary_in_usd'].median()
```

```python
# Medidas de dispersão
std_salary = df['salary_in_usd'].std()
min_salary = df['salary_in_usd'].min()
max_salary = df['salary_in_usd'].max()
```

```python
{
    "Média": mean_salary,
    "Mediana": median_salary,
    "Desvio Padrão": std_salary,
    "Mínimo": min_salary,
    "Máximo": max_salary
}
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histograma da distribuição geral dos salários
sns.histplot(df['salary_in_usd'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].axvline(mean_salary, color='red', linestyle='--', label=f'Média: {mean_salary:,.0f}')
axes[0].axvline(median_salary, color='green', linestyle='-', label=f'Mediana: {median_salary:,.0f}')
axes[0].set_title('Distribuição Geral dos Salários (USD)')
axes[0].set_xlabel('Salário em USD')
axes[0].set_ylabel('Frequência')
axes[0].legend()

# Boxplot por nível de experiência
sns.boxplot(data=df, x='experience_level', y='salary_in_usd', palette='pastel', ax=axes[1])
axes[1].set_title('Distribuição de Salários por Nível de Experiência')
axes[1].set_xlabel('Nível de Experiência')
axes[1].set_ylabel('Salário em USD')

plt.tight_layout()
plt.show()
```

```python
#Média de Salário por tamanha de Empresa
df.groupby('company_size')['salary_in_usd'].agg(['mean']).plot.bar()
```

```python
#Top 10 maiores Médias salariais por Cargo
top_10_salaries = (
    df.groupby('job_title')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_10_salaries
```

```python
#Top 10 maiores Médias salariais por País
top_10_salaries_for_cauntry = (
    df.groupby('company_location')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_10_salaries_for_cauntry
```

```python
# Correlação entre as variáveis Númericas

correlation_df = df[['salary_in_usd', 'work_year', 'years_of_experience']]

correlation_matrix = correlation_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f")
plt.title('Matriz de Correlação entre Salário, Ano e Tempo de Experiência')
plt.show()

correlation_matrix
```

🔍 Análise da Correlação das variáveis Númericas

✅ Existe alguma tendência de aumento salarial com os anos (work_year)?

Correlação positiva, mas fraca entre salary_in_usd e work_year (~0.18). Isso indica que os salários aumentaram levemente ao longo dos anos, mas não é uma tendência muito forte.

✅ Quanto maior o tempo de experiência, maior o salário?

Correlação moderada e positiva entre salary_in_usd e years_of_experience (~0.48). Isso sugere que, em geral, profissionais com mais experiência tendem a ganhar mais, embora não seja uma relação perfeita.
