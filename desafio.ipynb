```python
#Frequncia de Cargos
frequencia_cargos = df['job_title'].value_counts()

plt.figure(figsize=(12,5))

frequencia_cargos.plot(kind="bar", color="g")
plt.title("Frequncia Cargos")
plt.xlabel("Cargos")
plt.ylabel("Frequncia")

plt.tight_layout()
plt.show()
```

```python
#Frequncia de Experincia
frequencia_experiencia = df['experience_level'].value_counts()

plt.figure(figsize=(12,5))

frequencia_experiencia.plot(kind="bar", color="g")
plt.title("Frequncia Experincia")
plt.xlabel("Experincia")
plt.ylabel("Frequncia")

plt.tight_layout()
plt.show()
```

```python
#Frequncia de Tamanho de Empresa
frequencia_tam_empresa = df['company_size'].value_counts()

plt.figure(figsize=(12,5))

frequencia_tam_empresa.plot(kind="bar", color="g")
plt.title("Frequncia de Tamanho de Empresa")
plt.xlabel("Tamanho empresa")
plt.ylabel("Frequncia")

plt.tight_layout()
plt.show()
```

```python
# Medidas de tendncia central
mean_salary = df['salary_in_usd'].mean()
median_salary = df['salary_in_usd'].median()
```

```python
# Medidas de disperso
std_salary = df['salary_in_usd'].std()
min_salary = df['salary_in_usd'].min()
max_salary = df['salary_in_usd'].max()
```

```python
{
    "Mdia": mean_salary,
    "Mediana": median_salary,
    "Desvio Padro": std_salary,
    "Mnimo": min_salary,
    "Mximo": max_salary
}
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histograma da distribuio geral dos salrios
sns.histplot(df['salary_in_usd'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].axvline(mean_salary, color='red', linestyle='--', label=f'Mdia: {mean_salary:,.0f}')
axes[0].axvline(median_salary, color='green', linestyle='-', label=f'Mediana: {median_salary:,.0f}')
axes[0].set_title('Distribuio Geral dos Salrios (USD)')
axes[0].set_xlabel('Salrio em USD')
axes[0].set_ylabel('Frequncia')
axes[0].legend()

# Boxplot por nvel de experincia
sns.boxplot(data=df, x='experience_level', y='salary_in_usd', palette='pastel', ax=axes[1])
axes[1].set_title('Distribuio de Salrios por Nvel de Experincia')
axes[1].set_xlabel('Nvel de Experincia')
axes[1].set_ylabel('Salrio em USD')

plt.tight_layout()
plt.show()
```

```python
#Mdia de Salrio por tamanha de Empresa
df.groupby('company_size')['salary_in_usd'].agg(['mean']).plot.bar()
```

```python
#Top 10 maiores Mdias salariais por Cargo
top_10_salaries = (
    df.groupby('job_title')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_10_salaries
```

```python
#Top 10 maiores Mdias salariais por Pas
top_10_salaries_for_cauntry = (
    df.groupby('company_location')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_10_salaries_for_cauntry
```

```python
# Correlao entre as variveis Nmericas

correlation_df = df[['salary_in_usd', 'work_year', 'years_of_experience']]

correlation_matrix = correlation_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f")
plt.title('Matriz de Correlao entre Salrio, Ano e Tempo de Experincia')
plt.show()

correlation_matrix
```
